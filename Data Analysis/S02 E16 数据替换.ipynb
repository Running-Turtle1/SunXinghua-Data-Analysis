replace(): 替换
- 语法: str.replace(旧子串, 新子串, 替换次数)

```python
str = '你好, 世界! 你好,中国! 中国人 中国青岛'
```

```python
str.replace('中国', '东亚')
```

```python
str.replace('中国', '东亚', 1)
```

```python
str.replace('中国', '东亚', 100)
```
