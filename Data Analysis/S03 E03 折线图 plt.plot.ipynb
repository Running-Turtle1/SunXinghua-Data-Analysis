## 01. 折线图

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif']=['SimHei'] # 用黑体显示中文
```

```python
df = pd.read_excel('./repo2/08.折线图.xlsx')
```

```python
df,
```

## 01 单折线图

```python
plt.plot(
    df.时间,
    df.蔬菜,
    label = '蔬菜',
    color = 'r',
    marker = '*',
    # ms = 10 marker size
)
# for x,y in zip(df.时间,df.蔬菜):
#     plt.text(x, y+3, str(y), ha='center', va='center', fontsize=20, rotation=0)
plt.legend()
plt.tight_layout()
plt.show()
```

## 02 多折线图

```python
plt.plot(df.时间,df.蔬菜,label='蔬菜',color='r',marker='*',ms=5)
plt.plot(df.时间,df.水果,label='水果',color='g',marker='o',ms=5)
plt.plot(df.时间,df.食品,label='食品',color='b',marker='v',ms=5)
plt.plot(df.时间,df.用品,label='用品',color='y',marker='^',ms=5)
plt.legend(loc='upper left', bbox_to_anchor=(1.02, 1))
# plt.tight_layout()
plt.show()
```

给不同线段设置不同颜色

```python
plt.plot(df.时间[:3],df.蔬菜[:3],label='蔬菜',color='r',marker='*',ms=10)
plt.plot(df.时间[2:],df.蔬菜[2:],label='蔬菜',color='b',marker='*',ms=10)
plt.show()
```

## 02. 制作平均线

```python
import numpy as np
df = pd.read_excel(r'./repo2/09.折线与柱状组合图.xlsx')
# 柱图
plt.bar(df.班级,df.销售量,color='r',label='销售量',alpha=0.6)
# 显示图例
plt.legend()
# 平均线-折线图
平均线 = np.mean(df.销售量)
plt.axhline(y=平均线,color="blue",linestyle=':')
plt.show()
```
