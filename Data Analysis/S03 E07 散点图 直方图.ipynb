## 1 散点图

```python
import numpy as np
import matplotlib.pyplot as plt
k = 500
```

```python
x = np.random.rand(k) # 返回 [0-1] 之间均匀分布的随机样本
y = np.random.rand(k)
siz = np.random.rand(k) * 50
color = np.arctan2(x,y) # 生成每个点的颜色大小
plt.scatter(x, y, s = siz, c = color)
plt.colorbar() # 添加颜色栏
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# 显示所有列，同理：max_rows
pd.options.display.max_columns=None
plt.rcParams['font.sans-serif']=['SimHei'] # 用黑体显示中文
数据 = pd.read_excel(r'./repo2/16.散点图.xlsx')
plt.scatter(数据.身高,数据.体重,s=数据.身高,c=数据.身高)
plt.colorbar() # 添加颜色栏
plt.show()
```
