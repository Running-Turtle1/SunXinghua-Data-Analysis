## 1 map

```python
import pandas as pd
```

```python
path = './Repo/数据.xlsx'
```

```python
df = pd.read_excel(path)
```

```python
df
```

方法1：使用字典进行映射

```python
df['性别'] = df['性别'].map({
    '男': '先生',
    '女': '女士'
})
```

```python
df
```

方法2：使用函数

```python
def replace_(x):
    return '男的' if x == '先生' else '女的'
df['性别'] = df['性别'].map(replace_)
df
```

## 2 apply

apply与map区别：能够传入功能更为复杂的函数。

```python
path = './Repo/数据.xlsx'
df = pd.read_excel(path)
```

```python
def changeAge(x, delta):
    return x + delta
```

```python
df
```

```python
df['身高'] = df['身高'].apply(
    changeAge,
    args = (-10, )
)
```

```python
df
```

## 3 applymap

它的用法非常简单，会对DataFrame中的每个单元格执行指定函数的操作，虽然用途不如apply广泛，但在某些场合下还是比较有用：


```python
路径 = './Repo/数据2.xlsx'
数据 = pd.read_excel(路径)
数据2 = 数据.applymap(lambda x:"%.3f" % x)
print(数据2)
```
