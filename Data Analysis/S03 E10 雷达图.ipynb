```python
import numpy as np
import matplotlib.pyplot as plt
# 一个派是180度，极角是0.25*np.pi=45度，极轴20
plt.polar(0.25*np.pi,20,'ro')
plt.ylim(0,100)    #设置极轴的上下限
plt.show()
```

```python
# 给定多个极角和极轴的时候
import numpy as np
import matplotlib.pyplot as plt
角度 = np.array([0.25,0.75,1,1.5])
极轴 = [20,60,40,80]
plt.polar(角度*np.pi,极轴,'ro')
plt.ylim(0,100)
plt.show()
```

```python
# 将每个点连线
import numpy as np
import matplotlib.pyplot as plt
角度 = np.array([0.25,0.75,1,1.5])
极轴 = [20,60,40,80]
plt.polar(角度*np.pi,极轴,'ro-')
plt.ylim(0,100)
plt.show()
```

```python
# 构造一个极坐标点，和第一个点重叠
import numpy as np
import matplotlib.pyplot as plt
角度 = np.array([0.25,0.75,1,1.5,0.25])
极轴 = [20,60,40,80,20]
plt.polar(角度*np.pi,极轴,'ro-')
plt.ylim(0,100)
plt.show()
```
