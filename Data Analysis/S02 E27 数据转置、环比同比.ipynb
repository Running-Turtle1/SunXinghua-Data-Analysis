## 2 环比

$环比 = (本次数值 - 上次数值) / 上次数值$

```python
path = './Repo/环比.xlsx'
```

```python
df = pd.read_excel(path)
```

```python
df
```

```python
t = df.shift()
t
```

```python
df['环比'] = df.金额 - t.金额
```

## 3 同比

$同比 = (本期数值 - 上次本期数值) / 上次本期数值$

```python
df
```

```python
路径 = './Repo/同比.xlsx'
数据 = pd.read_excel(路径,'Sheet1')
年 = 数据['日期'].dt.year
数据2 = pd.pivot_table(数据,index='店号',values='金额',columns=年,aggfunc='sum')
同比 = (数据2[2019]-数据2[2018]) / 数据2[2018]
数据2['同比'] = 同比
print(数据2)
```
